```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

#Read in the data

df1 = pd.read_csv(r'C:\Users\rahulshankariyer\Downloads\Vaccinations against Covid Cases and Deaths by date.csv')
df2 = pd.read_csv(r'C:\Users\rahulshankariyer\Downloads\Vaccinations against Covid Cases and Deaths by country.csv')
```

```python
#Let's look at the date wise data

df1.head()
```

```python
#Let's look at the country wise data

df2.head()
```

```python
#Replacing null values with 0

df1.replace(to_replace = np.nan,value = 0)
```

```python
df2.replace(to_replace = np.nan,value = 0)
```

```python
#Vaccination vs Cases date wise

sns.regplot(x = 'total_vaccinations',y = 'total_cases',data = df1,scatter_kws = {"color" : "red"},line_kws = {"color" : "blue"})
```

```python
#Vaccination vs Deaths date wise

sns.regplot(x = 'total_vaccinations',y = 'total_deaths',data = df1,scatter_kws = {"color" : "red"},line_kws = {"color" : "blue"})
```

```python
#Vaccination vs Cases country wise

sns.regplot(x = 'new_vaccinations',y = 'new_cases',data = df2,scatter_kws = {"color" : "red"},line_kws = {"color" : "blue"})
```

```python
#Vaccination vs Deaths country wise

sns.regplot(x = 'new_vaccinations',y = 'new_deaths',data = df2,scatter_kws = {"color" : "red"},line_kws = {"color" : "blue"})
```

```python
#Correlation

correlation_matrix = df1.corr(method = 'pearson')
sns.heatmap(correlation_matrix,annot = True)
plt.title('Correlation Matrix for Covid Vaccinations against Cases & Deaths according to date')
plt.show()
```

```python
correlation_matrix = df2.corr(method = 'pearson')
sns.heatmap(correlation_matrix,annot = True)
plt.title('Correlation Matrix for Covid Vaccinations against Cases & Deaths according to country')
plt.show()
```

```python
#When we look at country wise data, the vaccinations have an equally high correlation with both cases and deaths
#When we look at date wise data, the vaccinations have a slightly higher correlation with cases than with deaths
#Vacccinations have a much higher correlation with both cases and deaths when looked at country wise than when looked at date wise
```
